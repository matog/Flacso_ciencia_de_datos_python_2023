```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
titanic = sns.load_dataset('titanic')
titanic.head()
```

## Ipywidgets

```python
#Creating the interactive dashboard
from ipywidgets import interact

@interact
def create_fare_plot(col = titanic.drop(['fare', 'age'], axis =1).columns): # Con drop marcamos que columnas NO aparecerán
    sns.barplot(data = titanic, x = col, y ='fare')
    plt.title(f'Bar plot de la tarifa media agrupado por  {col}')
```
